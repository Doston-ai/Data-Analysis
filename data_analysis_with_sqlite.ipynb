#💡Ishora: CREATE TABLE, INSERT


````python
#javobni shu yerda yozing.
cursor.execute("""
CREATE TABLE users2 (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name VARCHAR(60),
    birth_date INTEGER
)
""")
db.commit()
````

#**18)  AND and OR operators**

* `invoices` jadvalidan `BillingCountry` Germaniya yoki Fransiya boʻlgan va `Total` summasi 5 dan katta bo‘lgan kerakli qatorlarni tanlang.

#💡Ishora: AND, OR

````python
#javobni shu yerda yozing.
query = """
SELECT * FROM invoices WHERE (BillingCountry = 'Germany' or BillingCountry = 'France') AND Total>=5;
"""
invoices2 = pd.read_sql(query, db, index_col="InvoiceId")
invoices2
````

#**19) BETWEEN and LIKE**

````python
#javobni shu yerda yozing.
query = """
SELECT *FROM invoices WHERE Total BETWEEN 5 AND 15
"""
invoices3 = pd.read_sql(query, db, index_col="InvoiceId")
invoices3
````

````python
query = "SELECT * FROM customers WHERE FirstName LIKE 'A%';"
customers = pd.read_sql(query, db, index_col="CustomerId")
customers
````
